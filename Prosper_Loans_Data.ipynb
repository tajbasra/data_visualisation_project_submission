```python
import pandas as pd
import numpy as np

df = pd.read_csv('prosperLoanData.csv')

print df.head(5)
print df.dtypes
```

```python
df['ListingCreationDate'] = pd.to_datetime(df['ListingCreationDate'])
```

```python
gb = df.groupby(df.ListingCreationDate.dt.year)

#gb = df.groupby(df.ListingCreationDate.dt.to_period("M"))
```

```python
gb.describe()
```

```python
meanreturnsgrouping = gb['EstimatedReturn'].mean()
minreturnsgrouping = gb['EstimatedReturn'].min()
maxreturnsgrouping = gb['EstimatedReturn'].max()
countreturnsgrouping = gb['EstimatedReturn'].count()
```

```python
grouping = pd.DataFrame(data=dict(MeanReturns=meanreturnsgrouping, MinReturn=minreturnsgrouping, 
                       MaxReturn=maxreturnsgrouping, NumberOfListings=countreturnsgrouping))#, 
             #index=df.groupby(df.ListingCreationDate.dt.year).groups.keys())
```

```python
grouping.index.names = ['Year']
new_grouping = grouping.fillna(0)
print new_grouping
```

```python
new_grouping.to_csv('first_grouping.csv')
```

```python
grouping = grouping[pd.notnull(grouping['MaxReturn'])]
```

```python
newdf = pd.read_csv('grouping.csv')
del newdf['NumberOfListings']
```

```python
def gather( df, key, value, cols ):
    id_vars = [ col for col in df.columns if col not in cols ]
    id_values = cols
    var_name = key
    value_name = value
    return pd.melt( df, id_vars, id_values, var_name, value_name )
new_grouping = gather(newdf, 'Return', 'PercentReturn', ['MeanReturns', 'MinReturn', 'MaxReturn'])
```

```python
print new_grouping
```

```python
new_grouping.to_csv('new_grouping.csv')
```
